```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from tabulate import tabulate
# Group by 'release_year' and get titles for each year
grouped = df.groupby('release_year')['title'].apply(list).reset_index()

# Expand the grouped DataFrame to a more tabular format
expanded_rows = []

for index, row in grouped.iterrows():
    for title in row['title']:
        expanded_rows.append({'Year': row['release_year'], 'Title': title})

expanded_df = pd.DataFrame(expanded_rows)

# Print the first 20 rows using tabulate
print(tabulate(expanded_df.head(20), headers='keys', tablefmt='grid'))
```

```python
grouped=df.groupby('release_year')['title'].apply(list).reset_index()

# Print the result
for index, row in grouped.iterrows():
    # Print the release year
    print(f"Year: {row['release_year']}")
    # Print each title for the given year
    for title in row['title']:
        print(f" - {title}")
```

```python
sns.countplot(df['rating'])
sns.countplot(df['rating']).set_xticklabels(sns.countplot(df['rating']).get_xticklabels(), rotation=90, ha="right")
fig = plt.gcf()
fig.set_size_inches(13,13)
plt.title('Rating')
```

```python
labels = ['Movie', 'TV show']
size = df['type'].value_counts()
colors = plt.cm.Wistia(np.linspace(0, 1, 2))
explode = [0, 0.1]
plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(size,labels=labels, colors = colors, explode = explode, shadow = True, startangle = 90)
plt.title('Distribution of Type', fontsize = 25)
plt.legend()
plt.show()
```
